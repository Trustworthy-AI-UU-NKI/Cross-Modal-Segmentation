# Visualize Fake Images

```python
def vis_fake_im(orig, label, fake):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 4, 1)
    plt.title(f"True --> fake")
    plt.axis("off")
    plt.imshow(orig.squeeze(), cmap="gray")
    plt.subplot(1, 4, 2)
    plt.axis("off")
    plt.imshow(fake.squeeze(), cmap="gray")
    plt.subplot(1, 4, 3)
    plt.axis("off")
    plt.imshow((orig + label/3).squeeze())
    plt.subplot(1, 4, 4)
    plt.axis("off")
    plt.imshow((fake + label/3).squeeze())
    plt.show
```

```python
sl = 2
case = 1001
fold = 1

fake_ct = f"../data/other/fake_CT/run_fold_{fold}/images/case_{case}/slice_{sl}_0.nii.gz"
fake_ct_im = LoadImage()(fake_ct)
print(f"image data shape: {fake_ct_im.shape}")

label = f"../data/other/fake_CT/run_fold_{fold}/labels/case_{case}/slice_{sl}_0.nii.gz"
label_im = LoadImage()(label)
print(f"label data shape: {label_im.shape}")

img_mri_original = f"../data/other/fake_CT/run_fold_{fold}/original/case_{case}/slice_{sl}_0.nii.gz"
img_mri_original_im = LoadImage()(img_mri_original)
print(f"label data shape: {img_mri_original_im.shape}")

vis_fake_im(img_mri_original_im, label_im, fake_ct_im)
```
